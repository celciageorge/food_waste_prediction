```python
from scipy.stats import chi2_contingency
features_to_test = ['Category', 'Storage_Condition', 'Brand']
target = 'Waste_Risk'
for col in features_to_test:
    contingency_table = pd.crosstab(groceries[col], groceries[target])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    print(f"Relationship: {col} vs {target}")
    print(f"P-value: {p_value:.4f}")
    
    # Interpretation at 95% confidence level
    if p_value < 0.05:
        print(f"✅ Result: SIGNIFICANT. {col} has a strong impact on Waste Risk.")
    else:
        print(f"❌ Result: NOT SIGNIFICANT. {col} does not seem to affect Waste Risk.")
    print("-" * 30)
```

```python
cols_to_drop = ['Product_ID', 'Product_Name', 'Manufacture_Date', 'Expiry_Date', 'Country','Brand']

# Drop them from your dataframe
df = groceries.drop(columns=cols_to_drop)
df
```

```python
df.dtypes
```

```python
df1=pd.get_dummies(df[['Category','Storage_Condition']],drop_first=True,dtype=int)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2 = df2.drop(columns=['Category', 'Storage_Condition'])
df2
```

```python
df2.dtypes
```

```python
df2.shape
```

```python
# Force all expired items to be marked as Waste (1)
df2.loc[df2['Days_Left'] <=0, 'Waste_Risk'] = 1
```

```python
print(f"Total Expired Items: {len(df2[df2['Days_Left'] <= 0])}")
```

```python
x = df2.drop(columns=['Waste_Risk'])
x
```

```python
y=df2['Waste_Risk']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators=100, 
    max_depth=10,       # Limits how deep the trees can go (prevents memorization)
    min_samples_split=10, # Requires more data to make a decision
    random_state=42
)
model.fit(x_train, y_train)

```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score*100
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# 1. Get predictions for both sets
train_preds = model.predict(x_train)
test_preds = model.predict(x_test)

# 2. Calculate accuracy
from sklearn.metrics import accuracy_score
train_acc = accuracy_score(y_train, train_preds)
test_acc = accuracy_score(y_test, test_preds)

print(f"Training Accuracy: {train_acc * 100:.2f}%")
print(f"Testing Accuracy: {test_acc * 100:.2f}%")
```

```python
def predict_waste(quantity, shelf_life, days_left, price, category_dairy, category_grains, category_produce, category_protein, storage_refrigerated):
    # This matches the structure of your X_train
    data = [[quantity, shelf_life, days_left, price, category_dairy, category_grains, category_produce, category_protein, storage_refrigerated]]
    prediction = model.predict(data)
    
    if prediction[0] == 1:
        return "⚠️ High Risk of Waste! Consider a discount or donation."
    else:
        return "✅ Low Risk. Product is selling within a safe timeframe."

# Example Test: A Dairy item with only 1 day left
print(predict_waste(500, 7, 1, 50, 1, 0, 0, 0, 1))
```
